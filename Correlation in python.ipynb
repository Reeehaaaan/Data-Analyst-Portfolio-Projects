```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjusts the configuration of the plot which will be created

#read data  

df = pd.read_csv(r'C:\Users\DELL.WIN-EFQG1MH1O15\Desktop\Alex the analyst\Python\pp\moviess.csv')
df.head(20)
```

```python
df.head(20) # to display first 5 rows
```

```python
# checking for mising data

for col in df.columns:
    missing_per = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,missing_per))
```

```python
df.dtypes
```

```python
# changing data types
#df['budget'] = df['budget'].astype('int64')
#df['gross'] = df['gross'].astype('int64')

df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0).astype(int)
df['gross'] = pd.to_numeric(df['gross'], errors='coerce').fillna(0).astype(int) 
```

```python
#df['crct_year']= df['released'].astype(str).str[:4]

df['year_correct'] = df['released'].str.extract(pat = '([0-9]{4})').astype('Int64') # to extract year from released date
df.head(20)
```

```python
df.dtypes
```

```python
df['year_correct'] = df['year_correct'].astype(str)
```

```python
df.dtypes
```

```python
df.sort_values(by =['gross'], inplace =False, ascending = False)

df.head(20)
```

```python
#pd.set_option('display.max_rows', None)
df.head(20)
```

```python
# dropping duplicate
# df['company'].sort_values(ascending = False) # to check duplicate values
# df['company'].drop_duplicates().sort_values(ascending = False) --example
df.drop_duplicates() # drops duplicate from whole dataset

df.head(20)
```

```python
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget v/s Gross')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget of a film')
plt.show()
```

```python
sns.regplot(x ='budget', y='gross',data = df, scatter_kws={'color' : 'red'}, line_kws={'color' : 'black'})

plt.show()
```

```python
#looking at correlation, types are pearson, kendall, spearman, uses pearson by default
df.corr(method='pearson',numeric_only = True)
```

```python
correlation_matrix = df.corr(method='pearson',numeric_only = True)

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Budget for a film')
plt.show()
```

```python
#changing column with object type to numeric

df_numaric = df 

for col_mn in df_numaric.columns:
    if (df_numaric[col_mn].dtype == 'object'):
        df_numaric[col_mn] = df_numaric[col_mn].astype('category')
        df_numaric[col_mn] = df_numaric[col_mn].cat.codes

df_numaric.head(20)
```

```python
correlation_matrix = df_numaric.corr(method='pearson',numeric_only = True)

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Budget for a film')
plt.show()
```

```python
correlation_mat = df_numaric.corr()
corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
